```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# ================
# Global constants
# ================

pi = 3.1416

# integration time
ti = 0
tf = 13


# =======================
# Constant for particle 0
# =======================

par0_k = 1.0
par0_m = 2.0
par0_A = 2.0
par0_phi = 0.0
par0_w = math.sqrt(par0_k/par0_m)


# =======================
# Constant for particle 1
# =======================

par1_k = 4.0
par1_m = 2.0
par1_A = 2.0
par1_phi = 0.0
par1_w = math.sqrt(par1_k/par1_m)
```

```python
def position (w0, A0, phi0, t0):
    x = A0*math.cos(w0*t0+phi0)
    return x;

def velocity (w0, A0, phi0, t0):
    v = -A0*w0*math.sin(w0*t0+phi0)
    return v;

def acceleration (w0, A0, phi0, t0):
    a = -A0*w0*w0*math.cos(w0*t0+phi0)
    return a;

def energy_pot (m0, k0, x0):
    e_t = 0.5*k0*x0*x0
    return e_t;

def energy_kin (m0, k0, v0):
    e_t = 0.5*m0*v0*v0
    return e_t;
```

```python
# Defining/Initilizing vector for movement equations
par0_pos = []
par0_vel = []
par0_acc = []

t = np.arange(ti,tf,0.1)

for i in t:
    par0_pos.append(position(par0_w,par0_A,par0_phi,i))
    par0_vel.append(velocity(par0_w,par0_A,par0_phi,i))
    par0_acc.append(acceleration(par0_w,par0_A,par0_phi,i))

fig, ax = plt.subplots()

ax.plot(t,par0_pos,'r--',label="Position")
ax.plot(t,par0_vel,'b-*',label="Velocity")
ax.plot(t,par0_acc,'g^',label="Acceleration")

ax.xaxis.set_ticks(np.arange(ti, tf, 0.5))

plt.grid(True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.rcParams['figure.figsize'] = (12.0, 4.0)

plt.show()
```
