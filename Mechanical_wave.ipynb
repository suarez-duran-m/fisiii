```python
import matplotlib.pyplot as plt
import math
import numpy as np
import time
import threading
from IPython import display
import pylab as pl
```

```python
# ================
# Global constants
# ================

pi = 3.1416

# integration time
xrangei = 0
xrangef = 2
xrange = np.arange(xrangei,xrangef,.05)

# ===================
# Constant for System
# ===================

par0_vel = 100. # m/s
par0_A = 1. # m 
par0_lamb = 1 # m
par0_m = 1. # kg

par0_freq = par0_vel/par0_lamb
#par0_lamb = par0_vel/par0_freq
par0_k = 2*pi/par0_lamb
par0_w = 2*pi*par0_freq
```

```python
def position (w0, A0, k0, t0, x0):
    y= A0*math.cos(k0*x0 - w0*t0)
    return y;
```

```python
par0_pos = []

time0 = 0.
timef = .05
timerange = np.arange(time0,timef,.001)

for xx in xrange:
    par0_pos.append(position(par0_w,par0_A,par0_k,time0,xx))

plt.gca().cla()
plt.plot(xrange,par0_pos,'r-*',label="Position")
plt.grid(True)
plt.show()
```

```python
for tt in timerange:
    n = 0
    for xx in xrange:
        par0_pos[n] = position(par0_w,par0_A,par0_k,tt,xx)
        n = n+1

    plt.gca().cla()
    plt.plot(xrange,par0_pos,'b-*',label="Position")
    plt.grid(True)
    display.clear_output(wait=True)
    display.display(plt.gcf())                
    time.sleep(.5)
```
