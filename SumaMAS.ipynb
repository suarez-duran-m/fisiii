```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
ti = 0
tf = 50

A1=1.0
Phi1=0.7
W1=1.0

A2=1.0
Phi2=0.0
W2=1.02

Phi3=(Phi2-Phi1)
W3=(W2-W1)
```

```python
def position1 (w0, A0, phi0, t0):
    x=A0*math.cos(w0*t0+phi0)
    return x;
def position2 (w0, A0, phi0, t0):
    y = A0*math.cos(w0*t0+phi0)
    return y;
def position3 (w0, A0, A10, phi0, t0):
    z = math.sqrt(A0*A0+A10*A10+2*A0*A10*math.cos(w0*t0-phi0))
    return z;
```

```python
par0_pos1 = []
par0_pos2 = []
par0_pos3 = []
par0_pos4 = []

t = np.arange(ti,tf,0.1)

for i in t:    
    par0_pos1.append(position1(W1,A1,Phi1,i))    
    par0_pos2.append(position2(W2,A2,Phi2,i))    
    par0_pos3.append(position3(W3,A1,A2,Phi3,i))
    par0_pos4.append(position1(W1,A1,Phi1,i)+position2(W2,A2,Phi2,i)) 
    
fig, ax = plt.subplots()

#ax.plot(t,par0_pos1,'r--',label="Partícula 1")
#ax.plot(t,par0_pos2,'b-*',label="Partícula 2")
ax.plot(t,par0_pos3,'y*',label="Amplitud")
#ax.plot(t,par0_pos4,'g--',label="Suma MAS")

ax.xaxis.set_ticks(np.arange(ti, tf, 0.5))

plt.grid(True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.rcParams['figure.figsize'] = (12.0, 4.0)

plt.show()
```
