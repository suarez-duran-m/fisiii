```python
import matplotlib.pyplot as plt
import math
import numpy as np
import time
import threading
```

```python
# ================
# Global constants
# ================

pi = 3.1416

# integration time
ti = 0
tf = 100
t = np.arange(ti,tf,0.1)

# =======================
# Constant for particle 0
# =======================

par0_k = 9.0
par0_m = 2.0
par0_A = 2.0 # cm
par0_phi = 0.0 # rad
par0_w = math.sqrt(par0_k/par0_m)


# =======================
# Constant for particle 1
# =======================

par1_k = 18.0
par1_m = 2.0
par1_A = 2.0 # cm
par1_phi = 1.57 # rad
par1_w = math.sqrt(par1_k/par1_m)
```

```python
def position (w0, A0, phi0, t0):
    x = A0*math.cos(w0*t0+phi0)
    return x;

def velocity (w0, A0, phi0, t0):
    v = -A0*w0*math.sin(w0*t0+phi0)
    return v;

def acceleration (w0, A0, phi0, t0):
    a = -A0*w0*w0*math.cos(w0*t0+phi0)
    return a;

def energy_pot (m0, k0, x0):
    e_t = 0.5*k0*x0*x0
    return e_t;

def energy_kin (m0, k0, v0):
    e_t = 0.5*m0*v0*v0
    return e_t;
```

```python
# Defining/Initilizing vector for movement equations

# Particle 0
par0_pos = []
par0_vel = []
par0_acc = []

# Particle 1
par1_pos = []
par1_vel = []
par1_acc = []

# Sum

parS_pos = []
parS_vel = []
parS_acc = []
n = 0

for i in t:

    par0_pos.append(position(par0_w,par0_A,par0_phi,i))
    par0_vel.append(velocity(par0_w,par0_A,par0_phi,i))
    par0_acc.append(acceleration(par0_w,par0_A,par0_phi,i))

    par1_pos.append(position(par1_w, par1_A, par1_phi, i))
    par1_vel.append(velocity(par1_w, par1_A, par1_phi, i))
    par1_acc.append(acceleration(par1_w, par1_A, par1_phi, i))

    # Adding-up
    parS_pos.append( par0_pos[n] + par1_pos[n] )
    n += 1

fig, ax = plt.subplots()

ax.plot(t,parS_pos,'r--',label="Position")
#ax.plot(t,par1_vel,'b-*',label="Velocity")
#ax.plot(t,parS_acc,'g^',label="Acceleration")

ax.xaxis.set_ticks(np.arange(ti, tf, 10.))

plt.grid(True)
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.rcParams['figure.figsize'] = (12.0, 4.0)

plt.xlabel('Time [s]', fontsize=18)
plt.ylabel('Amplitude [cm]', fontsize=18)

plt.show()
```
