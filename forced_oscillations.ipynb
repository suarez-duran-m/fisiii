```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# ================
# Global constants
# ================

pi = 3.1416

# integration time
ti = 0
tf = 20
t = np.arange(ti,tf,0.1)

# =======================
# Constant for particle 0
# =======================

f_app = 50.0 # N
w_app = 1.0 # rad

par0_b = 0.05 # kg/s
par0_m = 1.0 # kg
par0_k = 25.0 # N/m
par0_w = math.sqrt(par0_k/par0_m)
par0_gamma = par0_b/par0_m
```

```python
def amplitude (f0, m0, w0, gamma0, w00):
    a = (f0/m0) / (w0*w0 * gamma0*gamma0 + (w00*w00 - w0*w0)**2)
    return a;
```

```python
par0_A = []
n = 0
for i in t:
    par0_A.append(amplitude(f_app, par0_m, par0_w, par0_gamma, i ))
    #print (i)
    print(par0_A[n])
    n += 1

print (par0_w) 
fig, ax = plt.subplots()

lw = 3.

ax.plot(t,par0_A,'r--',linewidth=lw)#,label='b ~ ')

delta_xticks = tf / 10.
ax.xaxis.set_ticks(np.arange(ti, tf, delta_xticks))

plt.grid(True)
#plt.legend(bbox_to_anchor=(1.1, 1.05),fontsize=24)
plt.rcParams['figure.figsize'] = (20.0, 10.0)

plt.xlabel('Time [s]', fontsize=24)
plt.ylabel('Amplitude [cm]', fontsize=24)

plt.xticks(fontsize = 24)
plt.yticks(fontsize = 24)

plt.show()
```
