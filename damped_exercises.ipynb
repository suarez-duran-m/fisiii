```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# ================
# Global constants
# ================

pi = 3.1416

# integration time
ti = 0
tf = 10
t = np.arange(ti,tf,0.1)

# =======================
# Constant for particle 0
# =======================

res0_k = 25.0 # N/m
par0_m = 0.05 # kg
par0_A = 1.0 # m
par0_phi = 0.0 # rad
par0_b = 0.022 # kg/s
```

```python
def position (A0, b0, phi0, m0, k0, t0):
    c0 = b0 / (2. * m0 )
    w0 = math.sqrt( k0 / m0 - ( b0*b0 ) / (4*m0*m0) )
    x = A0 * math.exp( -c0 * t0 ) * math.cos(w0 * t0 + phi0)
    return x;
```

```python
par0_pos = []
exp_term = []
par1_pos = []
par1_phi = pi/2

for i in t:    
    par0_pos.append(position( par0_A, par0_b, par0_phi, par0_m, res0_k, i ))
    par1_pos.append(position( par0_A, par0_b, par1_phi, par0_m, res0_k, i ))
    exp_term.append(math.exp((-par0_b/(2*par0_m))*i))
    
fig, ax = plt.subplots()

lw = 3.

ax.plot(t,par0_pos,'r--',linewidth=lw)#,label='b ~ ')
ax.plot(t,par1_pos,'g--',linewidth=lw)
#ax.plot(t,exp_term, 'b')

delta_xticks = tf / 10.
ax.xaxis.set_ticks(np.arange(ti, tf, delta_xticks))

plt.grid(True)
#plt.legend(bbox_to_anchor=(1.1, 1.05),fontsize=24)
plt.rcParams['figure.figsize'] = (20.0, 10.0)

plt.xlabel('Time [s]', fontsize=24)
plt.ylabel('Amplitude [cm]', fontsize=24)

plt.xticks(fontsize = 24)
plt.yticks(fontsize = 24)

plt.show()
```
